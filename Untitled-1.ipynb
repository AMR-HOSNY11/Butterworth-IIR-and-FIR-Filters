```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from scipy.signal import freqz

# --- Global Config ---
USE_TEST_SIGNAL = True        # Set to True if you don't have audio1.wav
TEST_SIGNAL_FREQ = 500         
TEST_SIGNAL_DUR  = 2.0         
fs = 8000

CUTOFF_HZ = 1000                
FILTER_ORDER = 1              # Higher order for better IIR roll-off
FIR_TAPS = 21                 
USE_IIR = False                # Toggle between True (IIR) and False (FIR)
FILTER_TYPE = "lowpass"        

ADD_NOISE = True               
NOISE_TYPE = "Hum"             
NOISE_LEVEL = 0.5              
NOISE_FREQUENCY = 60           # Standard power line hum

AUDIO_FILE = "audio1.wav"

# --- Utilities ---
def add_noise(audio, sr, noise_t, noise_lvl, noise_freq=60):
    if noise_t == 'None' or noise_lvl <= 0:
        return audio
    sig_std = np.sqrt(np.mean(audio**2))
    noise_amp = noise_lvl * sig_std
    t = np.arange(len(audio)) / sr
    noise = noise_amp * np.sin(2*np.pi*noise_freq*t)
    return audio + noise

# --- Butterworth IIR Design ---
def _butter_analog_poles(order):
    poles = []
    for k in range(1, order+1):
        theta = (2*k + order - 1) * np.pi / (2*order)
        poles.append(np.exp(1j*theta))
    return np.array(poles)

def design_butter_iir(filter_type, cutoff_hz, sr, order):
    omega_c = 2 * sr * np.tan(np.pi * cutoff_hz / sr)
    p_proto = _butter_analog_poles(order)
    p_analog = p_proto * omega_c
    a_s = np.real_if_close(np.poly(p_analog))
    
    # Bilinear transform
    z_poles = (1 + p_analog/(2*sr)) / (1 - p_analog/(2*sr))
    z_zeros = np.array([-1.0]*order) if filter_type == 'lowpass' else np.array([1.0]*order)

    a = np.real_if_close(np.poly(z_poles))
    b = np.real_if_close(np.poly(z_zeros))

    # Gain Normalization
    z_eval = 1.0 if filter_type == 'lowpass' else -1.0
    num = np.polyval(b, z_eval)
    den = np.polyval(a, z_eval)
    b = b * (np.real(den/num))
    return b, a, a_s, p_analog

# --- FIR Design ---
def design_fir_rectangular(filter_type, cutoff, sr, numtaps=101):
    omega = 2 * np.pi * cutoff / sr
    alpha = (numtaps - 1) / 2
    n = np.arange(numtaps)
    
    h_lp = (omega / np.pi) * np.sinc((omega / np.pi) * (n - alpha))
    
    if filter_type == "lowpass":
        h = h_lp / np.sum(h_lp)
    else: # highpass
        h = -h_lp
        h[int(alpha)] += 1
        h = h / np.sum(h * ((-1) ** n))
    return h, np.array([1.0])

# --- Filtering Engine ---
def apply_filter_direct(b, a, x):
    M, N = len(b)-1, len(a)-1
    y = np.zeros_like(x)
    for n in range(len(x)):
        acc = 0.0
        for i in range(M + 1):
            if n - i >= 0: acc += b[i] * x[n - i]
        for j in range(1, N + 1):
            if n - j >= 0: acc -= a[j] * y[n - j]
        y[n] = acc / a[0]
    return y

# --- Execution ---
if USE_TEST_SIGNAL:
    t = np.linspace(0, TEST_SIGNAL_DUR, int(fs * TEST_SIGNAL_DUR), endpoint=False)
    audio = 0.8 * np.sin(2*np.pi * TEST_SIGNAL_FREQ * t)
else:
    audio, fs = sf.read(AUDIO_FILE)
    if audio.ndim > 1: audio = audio.mean(axis=1)

audio_noisy = add_noise(audio, fs, NOISE_TYPE, NOISE_LEVEL, NOISE_FREQUENCY)

if USE_IIR:
    b, a, as_coeffs, p_an = design_butter_iir(FILTER_TYPE, CUTOFF_HZ, fs, FILTER_ORDER)
    filter_label = "IIR Butterworth"
else:
    b, a = design_fir_rectangular(FILTER_TYPE, CUTOFF_HZ, fs, FIR_TAPS)
    filter_label = "FIR Rectangular"

filtered = apply_filter_direct(b, a, audio_noisy)
filtered = filtered / (np.max(np.abs(filtered)) + 1e-12) # Normalize

# --- Visualization ---
w, H = freqz(b, a, worN=2048, fs=fs)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(w, 20*np.log10(np.abs(H) + 1e-9), 'cyan')
plt.title(f"{filter_label} Magnitude Response")
plt.axvline(CUTOFF_HZ, color='orange', linestyle='--')
plt.ylim([-60, 5])

plt.subplot(1, 2, 2)
plt.plot(w, np.angle(H), 'magenta')
plt.title(f"{filter_label} Phase Response")
plt.tight_layout()
plt.show()

# Z-Plane Plot
def plot_zplane(b, a):
    zeros, poles = np.roots(b), np.roots(a)
    
    plt.figure(figsize=(5, 5))
    circle = plt.Circle((0,0), 1, color='white', fill=False, linestyle='--')
    plt.gca().add_artist(circle)

    plt.scatter(zeros.real, zeros.imag, marker='o', edgecolors='blue', facecolors='none', label='Zeros')
    plt.scatter(poles.real, poles.imag, marker='x', color='red', label='Poles')

    plt.title("Z-Plane")
    plt.legend()
    plt.axis('equal')
    plt.grid(True, alpha=0.3)
    plt.show()

plot_zplane(b, a)
```

```python
# --- Calculate Frequency Response ---
# worN=2048 provides higher resolution for smoother plots
w, H = freqz(b, a, worN=2048, fs=fs)

# --- Visualization ---
plt.figure(figsize=(15, 6))

# 1. Magnitude Response (dB)
plt.subplot(1, 2, 1)
mag_db = 20 * np.log10(np.abs(H) + 1e-12) # Epsilon avoids log(0)
plt.plot(w, mag_db, color='cyan', linewidth=2)
plt.axvline(CUTOFF_HZ, color='red', linestyle='--', label='Cutoff')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.ylabel('Magnitude (dB)')
plt.xlabel('Frequency (Hz)')
plt.title(f"{FILTER_TYPE.capitalize()} ({'IIR' if USE_IIR else 'FIR'}) - Magnitude")
plt.ylim([max(-100, np.min(mag_db)), 5])
plt.legend()

# 2. Phase Response (Unwrapped Degrees)
plt.subplot(1, 2, 2)
# np.unwrap removes the 2*pi discontinuities
phase_rad = np.unwrap(np.angle(H))
phase_deg = np.rad2deg(phase_rad)

plt.plot(w, phase_deg, color='lime', linewidth=2)
plt.axvline(CUTOFF_HZ, color='red', linestyle='--')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.ylabel('Phase (degrees)')
plt.xlabel('Frequency (Hz)')
plt.title(f"{FILTER_TYPE.capitalize()} Phase Response (Unwrapped)")

plt.tight_layout()
plt.show()
```
